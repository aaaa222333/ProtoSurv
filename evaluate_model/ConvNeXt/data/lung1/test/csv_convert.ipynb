```python
import pandas as pd

# 读取CSV
df = pd.read_csv("/home/user/prognosis_lst/evaluate_model/ConvNeXt/data/lung1/test/final_complete_with_paths.csv")

# 1. 查看各列的数据类型
print("各列当前格式:")
print(df.dtypes)

# 2. 查看前几行数据
print("\n前3行数据:")
print(df.head(3))

# 3. 查看每列的唯一值（对分类变量有用）
print("\n每列唯一值示例:")
for col in df.columns:
    print(f"{col}: {df[col].unique()[:5]}")  # 显示前5个唯一值
```
