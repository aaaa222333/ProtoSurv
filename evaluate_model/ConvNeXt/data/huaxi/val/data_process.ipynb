```python
import pandas as pd

# 读取两个文件（没有表头）
df1 = pd.read_csv('/home/user/prognosis_lst/evaluate_model/ConvNeXt/data/huaxi/val/processed_rank_0_ori.csv', header=None)
df2 = pd.read_csv('/home/user/prognosis_lst/evaluate_model/ConvNeXt/data/huaxi/val/processed_rank_0_ori.csv', header=None)

# 添加列名方便操作
df1.columns = ['unique_id', 'path']
df2.columns = ['unique_id', 'path']

# 以第二列（unique_id）为唯一值合并
merged_df = pd.merge(df1, df2, on='unique_id', how='outer',suffixes=('_file1', '_file2'))

merged_df['path'] = merged_df['path_file1'].combine_first(merged_df['path_file2'])

# 保存结果

result = merged_df[['unique_id', 'path']].drop_duplicates(subset='unique_id', keep='first')
result.to_csv('/home/user/prognosis_lst/evaluate_model/ConvNeXt/data/huaxi/val/merged_output_ori.csv', index=True, header=True)

print(f"原始文件1行数: {len(df1)}")
print(f"原始文件2行数: {len(df2)}")
print(f"合并后行数: {len(result)}")
```

```python
import pandas as pd

# 1. 合并path文件
path_df = pd.read_csv("/home/user/prognosis_lst/evaluate_model/ConvNeXt/data/huaxi/val/merged_output_ori.csv")
# 2. 读取pred文件
pred_df = pd.read_csv('/home/user/prognosis_lst/evaluate_model/ConvNeXt/data/huaxi/val/val.csv')  # 如果有表头
# pred_df = pd.read_csv('pred.csv', header=None, names=['index', 'imgPath', 'pred'])  # 如果无表头

# 3. 左连接合并
result = pd.merge(
    path_df,
    pred_df[['imgPath', 'label']],
    left_on='path',
    right_on='imgPath',
    how='left'
).drop('imgPath', axis=1)

# 4. 找出缺失的
missing = result[result['label'].isna()]
print(f"缺失pred的记录数: {len(missing)}")
if len(missing) > 0:
    missing.to_csv('/home/user/prognosis_lst/evaluate_model/ConvNeXt/data/huaxi/val/missing_records.csv', index=False)
    print("缺失记录已保存")

result.to_csv('/home/user/prognosis_lst/evaluate_model/ConvNeXt/data/huaxi/val/final_complete_ori.csv', index=True)
print("完成！")
```
