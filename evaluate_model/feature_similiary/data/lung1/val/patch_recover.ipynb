```python
import pandas as pd
import os

# 读取CSV
df = pd.read_csv("final_complete.csv")

base_path = "/home/user/prognosis_lst/feature/lung1/val"

# 使用apply创建新列
df['feature_path'] = df['unique_id'].apply(lambda x: os.path.join(base_path, f"{x}.pt"))

# 查看结果
print(df.head())

# 保存（可以选择只保存需要的列）
df[['feature_path', 'label']].to_csv("final_complete_with_paths.csv", index=False)

print(f"已保存到: final_complete_with_paths.csv")
```
